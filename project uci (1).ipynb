```
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
```

```
from pyspark.ml.classification import LogisticRegression
lr = LogisticRegression()
```

```
featureCols=['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0','PAY_2',
 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4',
 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
```

```
assembler = VectorAssembler(inputCols=featureCols, outputCol="features")
```

```
assembled_d = assembler.transform(d)
```

```
display(assembled_d)
```

```
from pyspark.ml.feature import StandardScaler
standardScaler = StandardScaler(inputCol="features", outputCol="features_scaled")
```

```
scaled_d = standardScaler.fit(assembled_d).transform(assembled_d)
```

```
train_data, test_data = scaled_d.randomSplit([.8,.2], seed=100)
```

```
display(train_data)
```

```
from pyspark.ml.classification import LogisticRegression
```

```
lr=LogisticRegression(featuresCol='features_scaled',labelCol='default_payment_next_month',predictionCol='preddefault')
```

```
model=lr.fit(train_data)
```

```
predictions = model.transform(test_data)
```

```
predandlabels = predictions.select("preddefault", "default_payment_next_month")
```

```
train_summary=model.summary
train_summary.areaUnderROC
```

```
from pyspark.mllib.evaluation import BinaryClassificationMetrics
# Area under ROC curve
print("Area under ROC = %s" % train_summary.areaUnderROC)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator(labelCol="default_payment_next_month")
accuracy = evaluator.evaluate(predictions)
accuracy
```

```
#Decision Tree
from pyspark.ml.classification import DecisionTreeClassifier
```

```
dt=DecisionTreeClassifier(featuresCol='features_scaled',labelCol='default_payment_next_month',predictionCol='preddefault')
dtmodel=dt.fit(train_data)
```

```
dtpredictions = dtmodel.transform(test_data)
```

```
dtpredandlabels = dtpredictions.select("preddefault", "default_payment_next_month")
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
dtevaluator = MulticlassClassificationEvaluator(labelCol="default_payment_next_month",predictionCol='preddefault')
dtaccuracy = dtevaluator.evaluate(dtpredictions)
dtaccuracy
```

```
#RandomForest Classifier
from pyspark.ml.classification import RandomForestClassifier
```

```
rf=RandomForestClassifier(featuresCol='features_scaled',labelCol='default_payment_next_month',predictionCol='preddefault',numTrees=10)
```

```
rfmodel=rf.fit(train_data)
```

```
rfpredictions=rfmodel.transform(test_data)
```

```
rfpredandlabels = rfpredictions.select("preddefault", "default_payment_next_month")
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
rfevaluator = MulticlassClassificationEvaluator(labelCol="default_payment_next_month",predictionCol='preddefault')
rfaccuracy = rfevaluator.evaluate(rfpredictions)
rfaccuracy
```
